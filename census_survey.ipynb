```python
def albert_paul_split(dataframe, random_state=42):
    _columns = dataframe.columns
    _to_split = pd.DataFrame()
    _to_split['columns'] = _columns
    _albert = _to_split.sample(n=143, random_state=random_state)
    _albert = _albert.sort_index()
    _paul = _to_split.copy()
    _paul.drop(_albert.index, inplace=True)
    return _albert, _paul
```

```python
drop, keep = albert_paul_split(df)
```
